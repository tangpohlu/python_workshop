# NBS Stats. Web Scrapping

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

url = "https://stats.nba.com/alltime-leaders/"

def open_url_with_selenium_and_get_html_data(url):    
    driver = webdriver.Chrome()
    driver.get(url)
    html_data = driver.page_source
    driver.close()
    return html_data

data = open_url_with_selenium_and_get_html_data(url)

dfs = pd.read_html(data)[0]
print(dfs)

nba1 = r"C:\Users\PohLu\Documents\Personal\Python\NBA_Stats.xlsx"
export = dfs.to_excel(nba1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
dfs.groupby("Player")["GP"].sum().plot(kind="bar")
plt.ylabel('Points')
plt.xlabel('Player')
plt.title('Player Points')
```
